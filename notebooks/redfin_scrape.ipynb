```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin, urlparse

def scrape_redfin_property(url, output_html_path):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    }

    try:
        response = requests.get(url, headers=headers, timeout=30)
        response.raise_for_status()

        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Optional: rewrite src/href to absolute URLs if needed
        for tag in soup.find_all(["link", "script", "img"]):
            attr = "src" if tag.name in ["script", "img"] else "href"
            if tag.has_attr(attr):
                tag[attr] = urljoin(url, tag[attr])

        # Save the full HTML content
        with open(output_html_path, "w", encoding="utf-8") as f:
            f.write(soup.prettify())

        print(f"✅ HTML saved to: {output_html_path}")
        return True

    except requests.exceptions.RequestException as e:
        print(f"❌ Error fetching page: {e}")
        return False

# Example usage:
# scrape_redfin_property("https://www.redfin.com/CA/Fullerton/2334-E-Santa-Fe-Ave-92831/home/4909069", "output.html")
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import os

# Function to extract unit data from a single HTML file
def extract_unit_data_from_html(html_path):
    try:
        with open(html_path, 'r', encoding='utf-8') as f:
            soup = BeautifulSoup(f, 'html.parser')
    except Exception as e:
        print(f"Error reading {html_path}: {e}")
        return None

    # Get MLS ID from filename
    mls_id = os.path.splitext(os.path.basename(html_path))[0]
    
    # Initialize result dictionary with MLS ID
    result = {'MLS_ID': mls_id}
    
    # Extract property description from marketing remarks
    description = ""
    remarks_div = soup.find('div', {'data-rf-test-id': 'listingRemarks', 'id': 'marketing-remarks-scroll'})
    if remarks_div:
        # Find all text within the remarks div, including highlighted tags
        remarks_text = remarks_div.get_text(strip=True)
        if remarks_text:
            description = remarks_text
    
    result['property_description'] = description
    
    # Initialize columns for unit types 1-10
    for unit_num in range(1, 11):
        result[f'unit_type_{unit_num}_num_units'] = None
        result[f'unit_type_{unit_num}_beds'] = None
        result[f'unit_type_{unit_num}_baths'] = None
        result[f'unit_type_{unit_num}_rent'] = None
        result[f'unit_type_{unit_num}_actual_rent'] = None
        result[f'unit_type_{unit_num}_furnished'] = None

    # Find all unit type sections - check for both formats
    for i in range(1, 11):  # Check Unit Type 1 to 10
        # First check for "Unit Type X Information" format
        header = soup.find('div', string=lambda text: text and f'Unit Type {i} Information' in text)
        # If not found, check for "Unit X Information" format
        if not header:
            header = soup.find('div', string=lambda text: text and f'Unit {i} Information' in text)
        
        if not header:
            continue
        
        # Find the parent amenity-group div
        amenity_group = header.find_parent('div', class_='amenity-group')
        if not amenity_group:
            continue

        # Find all li elements within this specific amenity group only
        all_lis = amenity_group.find_all('li', class_='entryItem')
        
        for li in all_lis:
            entry_content = li.find('span', class_='entryItemContent')
            if not entry_content:
                continue
                
            # Get the full text content
            full_text = entry_content.get_text(strip=True)

            if 'Total # of Beds:' in full_text or '# of Beds:' in full_text:
                value_span = entry_content.find('span')
                if value_span:
                    result[f'unit_type_{i}_beds'] = value_span.get_text(strip=True)
            elif 'Total # of Baths:' in full_text or '# of Baths:' in full_text:
                value_span = entry_content.find('span')
                if value_span:
                    result[f'unit_type_{i}_baths'] = value_span.get_text(strip=True)
            elif 'Actual Rent:' in full_text:
                value_span = entry_content.find('span')
                if value_span:
                    rent_value = value_span.get_text(strip=True).replace('$', '').replace(',', '')
                    result[f'unit_type_{i}_actual_rent'] = rent_value
            elif 'Total Rent:' in full_text:
                value_span = entry_content.find('span')
                if value_span:
                    rent_value = value_span.get_text(strip=True).replace('$', '').replace(',', '')
                    result[f'unit_type_{i}_rent'] = rent_value
            elif 'Furnished:' in full_text:
                value_span = entry_content.find('span')
                if value_span:
                    result[f'unit_type_{i}_furnished'] = value_span.get_text(strip=True)
            elif 'Total # of Units:' in full_text or '# of Units:' in full_text:
                value_span = entry_content.find('span')
                if value_span:
                    result[f'unit_type_{i}_num_units'] = value_span.get_text(strip=True)

    return result

# Directory containing HTML files
scraped_html_dir = r"C:\Users\brand\Documents\projects\acres-ai\data\redfin\scraped_html"

# Get all HTML files
html_files = [f for f in os.listdir(scraped_html_dir) if f.endswith('.html')]

# Extract data from all files
all_data = []
for html_file in html_files:
    html_path = os.path.join(scraped_html_dir, html_file)
    unit_data = extract_unit_data_from_html(html_path)
    if unit_data:
        all_data.append(unit_data)

# Create combined DataFrame
combined_df = pd.DataFrame(all_data)

print(f"Processed {len(html_files)} HTML files")
print(f"Combined DataFrame shape: {combined_df.shape}")
```
